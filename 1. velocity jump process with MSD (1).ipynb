```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
num_steps = 1000
step_length = 1  # length of each run
turn_angle_std_dev = np.pi / 4  # standard deviation for turn angle distribution

# Initialize position and orientation
x_position = [0]
y_position = [0]
orientation = 0

# Initialize MSD
msd = []

# Simulate movement
for step in range(num_steps):
    orientation += np.random.normal(0, turn_angle_std_dev)
    x_position.append(x_position[-1] + step_length * np.cos(orientation))
    y_position.append(y_position[-1] + step_length * np.sin(orientation))
    
    # Calculate MSD
    msd.append(np.mean(np.array(x_position)**2 + np.array(y_position)**2))

# Plotting the path
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.plot(x_position, y_position)
plt.xlabel('X Position')
plt.ylabel('Y Position')
plt.title('Random Walk Model of Particle Movement')

plt.subplot(1, 2, 2)
plt.plot(msd)
plt.xlabel('Step')
plt.ylabel('Mean Square Displacement')
plt.title('Mean Square Displacement Over Time')

plt.tight_layout()
plt.show()
```

#### 4. Stimulate multiple particles 

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
num_steps = 1000
step_length = 1
turn_angle_std_dev = np.pi / 4
num_particles = 5  # Change this value to increase or decrease the number of particles

# Initialize position and orientation
x_positions = np.zeros((num_particles, num_steps+1))
y_positions = np.zeros((num_particles, num_steps+1))
orientations = np.zeros(num_particles)

# Initialize MSD for each particle
msd_particles = np.zeros((num_particles, num_steps+1))

# Simulate movement
for step in range(1, num_steps+1):
    orientations += np.random.normal(0, turn_angle_std_dev, num_particles)
    x_positions[:, step] = x_positions[:, step-1] + step_length * np.cos(orientations)
    y_positions[:, step] = y_positions[:, step-1] + step_length * np.sin(orientations)
    
    # Calculate MSD for each particle
    msd_particles[:, step] = x_positions[:, step]**2 + y_positions[:, step]**2

# Plotting the path
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
colors = ['b', 'g', 'r', 'c', 'm', 'y', 'k']  # Add more colors if you have more than 7 particles
for i in range(num_particles):
    plt.plot(x_positions[i], y_positions[i], color=colors[i % len(colors)], label=f'Particle {i+1}')
plt.xlabel('X Position')
plt.ylabel('Y Position')
plt.title('Random Walk Model of Multiple Particles')
plt.legend()

plt.subplot(1, 2, 2)
for i in range(num_particles):
    plt.plot(msd_particles[i], color=colors[i % len(colors)], label=f'Particle {i+1}')
plt.xlabel('Step')
plt.ylabel('Mean Square Displacement')
plt.title('Mean Square Displacement Over Time for Each Particle')
plt.legend()

plt.tight_layout()
plt.show()
```
